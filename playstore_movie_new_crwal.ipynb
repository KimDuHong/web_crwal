```python
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
```

## 영화 흥행순 정렬

```python
def 흥행순():
    path = os.getcwd() + "/chromedriver"
    broswer = webdriver.Chrome(path)
    show_list = []
    movie_url ="https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=%EB%B0%95%EC%8A%A4%EC%98%A4%ED%94%BC%EC%8A%A4&oquery=%EC%98%81%ED%99%94&tqi=hI0gtlp0J1ssscdgel0sssssskw-192872"
    broswer.get(movie_url)
    print("흥행순")
    tmp = broswer.find_element(By.ID, 'main_pack')
    all_movie = tmp.find_element(By.CLASS_NAME, "cm_content_wrap").text.split("\n")[2:-12]
    for i in range(len(all_movie)):
        if(i%3 == 0):
            show_list.append(all_movie[i] + "위 : " +all_movie[i+1] + " " +all_movie[i+2])
    for list in show_list:
        print(list)
    broswer.quit()
```

## 영화 평점순 정렬

```python
def 평점순():
    path = os.getcwd() + "/chromedriver"
    broswer = webdriver.Chrome(path)
    show_list = []
    movie_url ="https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=%EB%B0%95%EC%8A%A4%EC%98%A4%ED%94%BC%EC%8A%A4&oquery=%EC%98%81%ED%99%94&tqi=hI0gtlp0J1ssscdgel0sssssskw-192872"
    broswer.get(movie_url)
    broswer.find_element(By.XPATH,'//*[@id="main_pack"]/div[2]/div[2]/div/div/div[1]/div/div/ul/li[2]/a').send_keys(Keys.RETURN)
    print("평점순")
    tmp = broswer.find_element(By.ID, 'main_pack')
    all_movie = tmp.find_element(By.CLASS_NAME, "cm_content_wrap").text.split("\n")[2:-12]
    for i in range(len(all_movie)):
        if(i%3 == 0):
            show_list.append(all_movie[i] + "위 : " +all_movie[i+1] + " " +
                         all_movie[i+2])
    for list in show_list:
        print(list)
    broswer.quit()    
```

## 구글 뉴스 크롤링

```python
def get_news():
    path = os.getcwd() + "/chromedriver"
    broswer = webdriver.Chrome(path)
    print("구글 뉴스 기사\n")
    url ="https://www.google.com/search?q=%EA%B5%AC%EA%B8%80+%EB%89%B4%EC%8A%A4&rlz=1C5CHFA_enKR933KR933&source=lnms&tbm=nws&sa=X&ved=2ahUKEwjr1763nJ78AhWxUfUHHbLSAtAQ_AUoAXoECAEQAw&biw=1080&bih=1783&dpr=1"
    broswer.get(url)
    
    pages_nav = broswer.find_element(By.XPATH,'//*[@id="botstuff"]/div/div[2]/table/tbody/tr')
    
    pages = pages_nav.find_elements(By.TAG_NAME,'a')
    
    href_pages = []
    
    for page in pages:
        href_pages.append(page.get_attribute("href"))
        
    all_titles = []
    all_links = []
    
    for page in href_pages:
        url = page
        broswer.get(url)
        div = broswer.find_element(By.XPATH, '//*[@id="res"]')
        
        links = div.find_elements(By.TAG_NAME,'a')
        
        for link in links:
            all_links.append(link.get_attribute("href"))
            
        titles = div.find_elements(By.CLASS_NAME,'mCBkyc')
        
        for title in titles:
            all_titles.append(title.text)
            
        print(str(href_pages.index(page)+1) + "pages 크롤링중...")
        time.sleep(0.5)
    
    for i in range(len(all_titles)):
        print("제목 : " + all_titles[i] + "\n" + all_links[i] + "\n")
        

    broswer.quit()
```
