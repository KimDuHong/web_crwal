```python
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
```

```python
def get_journal():
    delay = 0.5
    long_delay = 1.5
    path = os.getcwd() + "/chromedriver"
    driver = webdriver.Chrome(path)
    
    base_url ="https://www.dbpia.co.kr/search/topSearch?searchOption=all&query="
    word = "아"
    driver.get(base_url + word)

    
    buttons = []
    
    while(len(buttons) == 0):
        print("while")
        time.sleep(delay)
        buttons = driver.find_element(By.ID,"pageList").find_elements(By.CLASS_NAME,"dpPaging__link")
     
    
    for i in range(10):
        
        div = driver.find_element(By.ID,"searchResultList")
        time.sleep(delay)
        lists = div.find_elements(By.CLASS_NAME,"thesis")

        links = [] ## 이동용 링크
        titles = [] ## 제목 
        authors = [] ## 저자
        
        for list in lists:
            check = list.find_element(By.TAG_NAME,"input").get_attribute("class")

            if(check.find("disabled") != -1):
                continue
            else:
                links.append(list.find_element(By.CLASS_NAME,"thesis__link"))

                title = list.find_element(By.CLASS_NAME,"thesis__tit").text
                titles.append(title)

                author = list.find_element(By.CLASS_NAME,"thesis__author").text
                authors.append(author)

        contents = [] ## 내용이 담긴 배열

        for link in links:
            link.click()
            time.sleep(delay)
            driver.switch_to.window(driver.window_handles[1] )
            time.sleep(delay)
            print(titles[links.index(link)])

            try:
                driver.find_element(By.CLASS_NAME,"thesisDetail__abstract").find_element(By.CLASS_NAME,"noData")
                contents.append("등록된 초록이 없습니다.")
            except:
                time.sleep(delay)
                contents.append(driver.find_element(By.CLASS_NAME,"abstractTxt").text)

            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            
        
        buttons = driver.find_element(By.ID,"pageList").find_elements(By.CLASS_NAME,"dpPaging__link")
        current_page = 0
        for button in buttons:
            if button.get_attribute("class").find("current") != -1:
                current_page = buttons.index(button)
                break
        
        if current_page != len(buttons):
            buttons[current_page+1].send_keys(Keys.ENTER)
            time.sleep(delay)
        
        
       
```

```python
get_journal()
```
